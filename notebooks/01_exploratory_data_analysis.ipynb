```python
# ==========================================
# CELL 1: Import Libraries & Load Data (Days 3-4)
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load a built-in dataset for speed
df = sns.load_dataset('titanic')
print("--- Dataset Loaded Successfully ---")
print(df.head())

# Save raw data to your structure
df.to_csv('../data/raw/titanic_raw.csv', index=False)

# ==========================================
# CELL 2: Data Cleaning & Preprocessing (Days 5-6)
# ==========================================
print("\n--- Missing Values Before Cleaning ---")
print(df.isnull().sum())

# Handle missing values strategically
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df.drop(columns=['deck', 'embark_town'], inplace=True, errors='ignore') # Drop highly missing columns

# Save clean data to your structure
df.to_csv('../data/processed/titanic_clean.csv', index=False)
print("\n--- Data Cleaning Complete: Cleaned Dataset Saved ---")

# ==========================================
# CELL 3: Exploratory Data Analysis (Days 7-8)
# ==========================================
# 1. Descriptive Statistics
print("\n--- Statistical Summary ---")
print(df.describe())

# 2. Visualizations
plt.figure(figsize=(12, 5))

# Plot 1: Survival Rate Distribution (Univariate)
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='survived')
plt.title('Distribution of Survival')

# Plot 2: Age vs Fare by Survival Status (Bivariate)
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='age', y='fare', hue='survived', alpha=0.6)
plt.title('Age vs Fare Analysis')

plt.tight_layout()
plt.savefig('../reports/eda_plots.png')
plt.show()

# Plot 3: Correlation Matrix
plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('../reports/correlation_matrix.png')
plt.show()
```
