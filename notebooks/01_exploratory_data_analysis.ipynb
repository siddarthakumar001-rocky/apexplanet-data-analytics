```python
# ==========================================
# CELL 1: Import Libraries & Load Data
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Using the absolute path to guarantee it finds the file
df = pd.read_csv(r'D:\apexplanet\data\raw\fifa_world_cup_2026_player_performance-selected-columns.csv')
print("--- FIFA 2026 Dataset Loaded Successfully ---")
print(df.head())

# ==========================================
# CELL 2: Data Cleaning & Preprocessing
# ==========================================
print("\n--- Missing Values Before Cleaning ---")
print(df.isnull().sum())

# Drop any rows with missing data
df = df.dropna()

# Save the clean version using absolute path
df.to_csv(r'D:\apexplanet\data\processed\fifa_2026_clean.csv', index=False)
print("\n--- Data Cleaning Complete: Cleaned Dataset Saved ---")
print("Total rows ready for analysis:", len(df))

# ==========================================
# CELL 3: Exploratory Data Analysis (EDA)
# ==========================================
# 1. Descriptive Statistics
print("\n--- Statistical Summary ---")
print(df.describe())

# 2. Visualizations
plt.figure(figsize=(16, 5))

# Plot 1: Player Positions
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='position', palette='viridis')
plt.title('Distribution of Player Positions')
plt.xlabel('Position')
plt.ylabel('Count')

# Plot 2: Age vs Height
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='age', y='height_cm', hue='preferred_foot', alpha=0.6)
plt.title('Player Age vs Height (cm)')
plt.xlabel('Age')
plt.ylabel('Height (cm)')

# Plot 3: Preferred Foot
plt.subplot(1, 3, 3)
sns.countplot(data=df, x='preferred_foot', palette='Set2')
plt.title('Preferred Foot Comparison')
plt.xlabel('Foot')
plt.ylabel('Count')

plt.tight_layout()

# Save the plot using absolute path
plt.savefig(r'D:\apexplanet\reports\fifa_eda_plots.png')
plt.show()
```
